# Intialise new DataFrame for each performance metrics

```python
explored_df = pd.DataFrame({'DFS':[],'BFS':[],"GBFS":[],"A*":[],"Djikstra":[], "IDA*":[]})
path_len_df = pd.DataFrame({'DFS':[],'BFS':[],"GBFS":[],"A*":[],"Djikstra":[], "IDA*":[]})
time_df = pd.DataFrame({'DFS':[],'BFS':[],"GBFS":[],"A*":[],"Djikstra":[], "IDA*":[]})

explored_df
```

```python
import time

def evaluate_algo(maze, algo):
    st = time.time() # get the start time
    if algo == "DFS":
        explored, path = DFS(maze)
    elif algo == "BFS":
        explored, path = BFS(maze)
    elif algo == "GBFS":
        explored, path = GBFS(maze)
    elif algo == "A*":
        explored, path = A_star(maze)
    elif algo == "Djikstra":
        explored, path = dijkstra(maze)
    elif algo == "IDA*":
        explored, path = ida_star(maze)
    et = time.time() # get the end time

    elapsed = (et - st)*1000

    return explored, path, elapsed
```

# Running the automated test

```python
for i in range(100):
    width = random.randint(30, 60)
    height = random.randint(30, 60)
    maze = generate_maze(width, height)

    explored_df = explored_df.append({'DFS':0,'BFS':0,"GBFS":0,"A*":0,"Djikstra":0,"IDA*":0}, ignore_index=True)
    path_len_df = path_len_df.append({'DFS':0,'BFS':0,"GBFS":0,"A*":0,"Djikstra":0,"IDA*":0}, ignore_index=True)
    time_df = time_df.append({'DFS':0,'BFS':0,"GBFS":0,"A*":0,"Djikstra":0,"IDA*":0}, ignore_index=True)
    
    for algo in ["DFS", "BFS", "GBFS", "A*", "Djikstra", "IDA*"]:
        explored, path, elapsed = evaluate_algo(maze, algo)

        if explored is not None:
            explored_df.loc[i, algo] = len(explored)
            path_len_df.loc[i, algo] = len(path)
        else:
            explored_df.loc[i, algo] = explored
            path_len_df.loc[i, algo] = path
        time_df.loc[i, algo] = elapsed

    print("Done test case", i)
```
